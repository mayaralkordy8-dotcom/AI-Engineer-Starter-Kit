```python
import requests

def get_random_quote():
    url = "https://zenquotes.io/api/random"
    try:
        # Send a GET request to the server
        response = requests.get(url)
        response.raise_for_status() # Ensure the request was successful

        # Convert response to JSON
        data = response.json()

        # Extract the quote text and its author
        quote = data[0]['q']
        author = data[0]['a']
        return quote, author
    except Exception as e:
        print(f"An error occurred while fetching the quote: {e}")
        # Fallback quote in case of internet or server issues
        return "The only limit to our realization of tomorrow will be our doubts of today.", "Franklin D. Roosevelt"

# Run the function and save the results in variables
quote_text, author_name = get_random_quote()

print(f"Quote: \"{quote_text}\"")
print(f"Author: - {author_name}")
```

```python
from transformers import pipeline

# Initialize the Sentiment Analysis pipeline
print("Loading AI model...")
classifier = pipeline("sentiment-analysis")

# Analyze the quote fetched in the previous step
print("\nAnalyzing sentiment...")
result = classifier(quote_text)[0]

# Extract classification label and confidence score
label = result['label']
score = result['score']

# Print the final results in a clean format
print("\n" + "="*40)
print("SENTIMENT ANALYSIS RESULT")
print("="*40)
print(f"Quote: \"{quote_text}\"")
print(f"Sentiment: {label}")
print(f"Confidence Score: {score:.4f}")
print("="*40)
```
